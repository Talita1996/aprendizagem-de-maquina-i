```python
knn = knn.drop(columns=['Start Time','Run ID','Source Type','Source Name','User','Status'])
dt = dt.drop(columns=['Start Time','Run ID','Source Type','Source Name','User','Status'])
mlp = mlp.drop(columns=['Start Time','Run ID','Source Type','Source Name','User','Status'])
nb = nb.drop(columns=['Start Time','Run ID','Source Type','Source Name','User','Status'])
svm = svm.drop(columns=['Start Time','Run ID','Source Type','Source Name','User','Status'])
rf = rf.drop(columns=['Start Time','Run ID','Source Type','Source Name','User','Status'])
lr = lr.drop(columns=['Start Time','Run ID','Source Type','Source Name','User','Status'])
```

```python
def selected_metric(row):
    name = row['Name']
    if 'AUC' in name: return 'Área sob a curva ROC'
    elif 'Fbeta' in name: return f'F{chr(946)}-Score ({chr(946)} = 2)'
    elif 'F1' in name: return 'F1-Score'
    elif 'REVO' in name: return 'Revocação'
    elif 'PREC' in name: return 'Precisão'
    else: return 'Acurácia'
```

```python
knn['selected_metric'] = knn.apply(lambda row: selected_metric(row), axis=1)
dt['selected_metric'] = dt.apply(lambda row: selected_metric(row), axis=1)
mlp['selected_metric'] = mlp.apply(lambda row: selected_metric(row), axis=1)
nb['selected_metric'] = nb.apply(lambda row: selected_metric(row), axis=1)
svm['selected_metric'] = svm.apply(lambda row: selected_metric(row), axis=1)
rf['selected_metric'] = rf.apply(lambda row: selected_metric(row), axis=1)
lr['selected_metric'] = lr.apply(lambda row: selected_metric(row), axis=1)
```

```python
nb.head(1)
```

```python
def selected_metric_value(row, model):
    name = row['Name'].replace(model, 'best_train')
    return row[name]
```

```python
knn['selected_metric_best_train'] = knn.apply(lambda row: selected_metric_value(row, 'KNN'), axis=1)
dt['selected_metric_best_train'] = dt.apply(lambda row: selected_metric_value(row, 'DecisionTree'), axis=1)
mlp['selected_metric_best_train'] = mlp.apply(lambda row: selected_metric_value(row, 'MLP'), axis=1)
nb['selected_metric_best_train'] = nb.apply(lambda row: selected_metric_value(row, 'NaiveBayes'), axis=1)
svm['selected_metric_best_train'] = svm.apply(lambda row: selected_metric_value(row, 'SVM'), axis=1)
rf['selected_metric_best_train'] = rf.apply(lambda row: selected_metric_value(row, 'RandomForest'), axis=1)
lr['selected_metric_best_train'] = lr.apply(lambda row: selected_metric_value(row, 'LogisticRegression'), axis=1)
```

```python
knn = knn.drop(columns=['best_train_ACURACIA','best_train_F1','best_train_Fbeta_2','best_train_PRECISAO','best_train_REVOCACAO','best_train_ROC_AUC'])
dt = dt.drop(columns=['best_train_ACURACIA','best_train_F1','best_train_Fbeta_2','best_train_PRECISAO','best_train_REVOCACAO','best_train_ROC_AUC'])
mlp = mlp.drop(columns=['best_train_ACURACIA','best_train_F1','best_train_Fbeta_2','best_train_PRECISAO','best_train_REVOCACAO','best_train_ROC_AUC'])
nb = nb.drop(columns=['best_train_ACURACIA','best_train_F1','best_train_Fbeta_2','best_train_PRECISAO','best_train_REVOCACAO','best_train_ROC_AUC'])
svm = svm.drop(columns=['best_train_ACURACIA','best_train_F1','best_train_Fbeta_2','best_train_PRECISAO','best_train_REVOCACAO','best_train_ROC_AUC'])
rf = rf.drop(columns=['best_train_ACURACIA','best_train_F1','best_train_Fbeta_2','best_train_PRECISAO','best_train_REVOCACAO','best_train_ROC_AUC'])
lr = lr.drop(columns=['best_train_ACURACIA','best_train_F1','best_train_Fbeta_2','best_train_PRECISAO','best_train_REVOCACAO','best_train_ROC_AUC'])
```

```python
knn['algorithm'] = knn.apply(lambda row: 'kNN', axis=1)
dt['algorithm'] = dt.apply(lambda row: 'Decision Tree (CART)', axis=1)
mlp['algorithm'] = mlp.apply(lambda row: 'MLP', axis=1)
nb['algorithm'] = nb.apply(lambda row: 'Naive Bayes', axis=1)
svm['algorithm'] = svm.apply(lambda row: 'SVM', axis=1)
rf['algorithm'] = rf.apply(lambda row: 'Random Forest', axis=1)
lr['algorithm'] = lr.apply(lambda row: 'Logistic Regression', axis=1)
```

```python
def get_config(row, model):
    result = ''
    if model == 'KNN':
        result = f"n_neighbors = {row['n_neighbors']}; weights = {row['weights']}"
    elif model == 'DT':
        result = f"criterion = {row['criterion']}; max_depth = {row['max_depth']}"
    elif model == 'MLP':
        result = f"activation = {row['activation']}; alpha = {row['alpha']}; hidden_layer_sizes = {row['hidden_layer_sizes']}; learning_rate = {row['learning_rate']}; solver = {row['solver']}"
    elif model == 'SVM':
        result = f"c = {row['C']}; gamma = {row['gamma']}; kernel = {row['kernel']}"
    elif model == 'RF':
        result = f"criterion = {row['criterion']}; max_depth = {row['max_depth']}; n_estimators = {row['n_estimators']}"
    elif model == 'LR':
        result = f"c = {row['C']}; solver = {row['solver']}; max_iter = {row['max_iter']}; penalty = {row['penalty']}"
    else:
        result = f"var_smoothing = {row['var_smoothing']}"
    return result
```

```python
knn['best_params'] = knn.apply(lambda row: get_config(row, 'KNN'), axis=1)
dt['best_params'] = dt.apply(lambda row: get_config(row, 'DT'), axis=1)
mlp['best_params'] = mlp.apply(lambda row: get_config(row, 'MLP'), axis=1)
nb['best_params'] = nb.apply(lambda row: get_config(row, 'NB'), axis=1)
svm['best_params'] = svm.apply(lambda row: get_config(row, 'SVM'), axis=1)
rf['best_params'] = rf.apply(lambda row: get_config(row, 'RF'), axis=1)
lr['best_params'] = lr.apply(lambda row: get_config(row, 'LR'), axis=1)
```

```python
knn_cols = {col : col.lower().replace('test_', '') for col in knn.columns.to_list()}
knn = knn.rename(columns=knn_cols)
knn = knn.drop(columns=['n_neighbors', 'weights'])

dt_cols = {col : col.lower().replace('test_', '') for col in dt.columns.to_list()}
dt = dt.rename(columns=dt_cols)
dt = dt.drop(columns=['criterion','max_depth'])

mlp_cols = {col : col.lower().replace('test_', '') for col in mlp.columns.to_list()}
mlp = mlp.rename(columns=mlp_cols)
mlp = mlp.drop(columns=['activation','alpha','hidden_layer_sizes','learning_rate','solver'])

nb_cols = {col : col.lower().replace('test_', '') for col in nb.columns.to_list()}
nb = nb.rename(columns=nb_cols)
nb = nb.drop(columns=['var_smoothing'])

svm_cols = {col : col.lower().replace('test_', '') for col in svm.columns.to_list()}
svm = svm.rename(columns=svm_cols)
svm = svm.drop(columns=['c', 'gamma', 'kernel'])

rf_cols = {col : col.lower().replace('test_', '') for col in rf.columns.to_list()}
rf = rf.rename(columns=rf_cols)
rf = rf.drop(columns=['criterion', 'max_depth', 'n_estimators'])

lr_cols = {col : col.lower().replace('test_', '') for col in lr.columns.to_list()}
lr = lr.rename(columns=lr_cols)
lr = lr.drop(columns=['c', 'solver', 'max_iter', 'penalty'])
```

```python
knn = knn[['algorithm','selected_metric','selected_metric_best_train','best_params','acuracia','precisao','revocacao','f1','fbeta_2','roc_auc']].copy()
dt = dt[['algorithm','selected_metric','selected_metric_best_train','best_params','acuracia','precisao','revocacao','f1','fbeta_2','roc_auc']].copy()
mlp = mlp[['algorithm','selected_metric','selected_metric_best_train','best_params','acuracia','precisao','revocacao','f1','fbeta_2','roc_auc']].copy()
nb = nb[['algorithm','selected_metric','selected_metric_best_train','best_params','acuracia','precisao','revocacao','f1','fbeta_2','roc_auc']].copy()
svm = svm[['algorithm','selected_metric','selected_metric_best_train','best_params','acuracia','precisao','revocacao','f1','fbeta_2','roc_auc']].copy()
rf = rf[['algorithm','selected_metric','selected_metric_best_train','best_params','acuracia','precisao','revocacao','f1','fbeta_2','roc_auc']].copy()
lr = lr[['algorithm','selected_metric','selected_metric_best_train','best_params','acuracia','precisao','revocacao','f1','fbeta_2','roc_auc']].copy()
```
